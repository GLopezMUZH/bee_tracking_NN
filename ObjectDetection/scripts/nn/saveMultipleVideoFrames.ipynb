```
import cv2
import numpy as np
import os
import glob
from time import gmtime, strftime  
```

```
def getVideosInDir(dir_path):
  print("getimagesInDir", dir_path)
  image_list = []
  for filename in glob.glob(dir_path + '/*.mp4'):
    print("vid", filename)
    image_list.append(filename)

  return image_list
```

```
dir_path_videos = '/content/gdrive/My Drive/beeTracking_img_labeling/Froh23'
dir_path_frames_train = '/content/gdrive/My Drive/beeTracking_img_labeling/Froh23/train'
dir_path_frames_test = '/content/gdrive/My Drive/beeTracking_img_labeling/Froh23/test'

hive = 'Froh23'
# 20 vids at 2 mins per video * 60 seconds * 25 frames per second
nr_frames = (20 * 2 * 60 * 25)
frequency_saving_test = 500
frequency_saving_train =  501
```

```
print(strftime("%Y-%m-%d %H:%M:%S", gmtime()))
videos = glob.glob(dir_path_videos+"/*.mp4")
curr_video = 1

for video in videos:
  print(video)
  print(strftime("%Y-%m-%d %H:%M:%S", gmtime()))
  curr_video += 1

  # Playing video from file:
  cap = cv2.VideoCapture(video)

  try:
    if not os.path.exists(hive+'/frames'):
      os.makedirs(hive+'/frames')
  except OSError:
    print('Error: Creating directory of data')

  currentFrame = 0
  while(True):
    currentFrame += 1
    ret, frame = cap.read()

    # train
    if currentFrame % frequency_saving_train == 0:
      # Saves image of the current frame in jpg file
      video_name = str(video.split('/')[-1].replace('.mp4',''))
      name = dir_path_frames_train+'/'+video_name+'_frame_'+ str(currentFrame) + '.jpg'
      print('Creating...' + name)
      try:
        cv2.imwrite(name, frame)
      except:
        print('Error', currentFrame)
        break

    """ 
    # test
    if currentFrame % frequency_saving_test == 0:
      # Saves image of the current frame in jpg file
      video_name = str(video.split('/')[-1].replace('.MOV',''))
      name = dir_path_frames_test+'/'+video_name+'_frame_'+ str(currentFrame) + '.jpg'
      print('Creating...' + name)
      try:
        cv2.imwrite(name, frame)
      except:
        print('Error', currentFrame)
        break
    """

    if cv2.waitKey(25) & 0xFF == ord('q'):
      break

  # To stop duplicate images
  print("Next video: ", curr_video)

  # When everything done, release the capture
  cap.release()
  cv2.destroyAllWindows()
  print("End")
```
