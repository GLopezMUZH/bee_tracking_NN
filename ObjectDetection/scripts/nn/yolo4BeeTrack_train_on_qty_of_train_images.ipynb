Get current status of training set

```
import os.path
from numpy import genfromtxt


status = ['new', 'started','best']

training_type = 'on_qty_of_train_images'
cfg_files_folder =  "/content/gdrive/My Drive/yolo_models/cfg_files/yolo_train_files/on_qty_of_train_images"

file_name = training_type + '.csv'
file_path = cfg_files_folder + '/' + file_name
print(file_path)

if not os.path.isfile(file_path):
  # create list file of models to train
  print("create file models_train_status")
  list_file = open(file_path, 'w')
  files = []
  files = os.listdir(cfg_files_folder)
  files = [f for f in files if f[len(f)-4:len(f)] == 'data']
  list_file.write("cfg_file,train_status,train_start_time,train_end_time\n")

  for file in files:
    t = strftime("%Y-%m-%d %H:%M:%S", gmtime())
    st = cfg_files_folder + '/' + file + ',new,' + t + ','+ t 
    list_file.write("%s\n" % st)
  list_file.close()


models_train_status = genfromtxt(file_path, delimiter=',',dtype="U255,U8,U20,U20",encoding="utf8", names = True)

print(models_train_status)
```

```
 f = models_train_status[:1]['cfg_file'][0]
 print(f[len(f)-4:len(f)])
 print(f[len(cfg_files_folder)+1:len(f)-5])
```

## Train according to status file

```
for mts in models_train_status:
  print(len(mts['train_status']), end= '')
  print(mts['train_status'])
  #print(mts)
  #mts['train_status'] = 'new'
```

```
for mts in models_train_status:
  f = mts['cfg_file']
  dfg_file_path = ""+ f +""
  print("Train: ", dfg_file_path)
  print("Status: ", mts['train_status'])

  backup_folder_name = f[len(cfg_files_folder)+1:len(f)-5]
  backup_folder_path = '/content/gdrive/My Drive/yolo_models/backup_yolo4/on_qty_of_train_images/' + backup_folder_name + '/'
  if not os.path.exists(backup_folder_path):
    print("created backup folder")
    os.makedirs(backup_folder_path)

  if mts['train_status'] == 'new':
    # from scratch 
    print("Start time: ", mts['train_start_time'])
    mts['train_status'] = 'started'
    mts['train_start_time'] = strftime("%Y-%m-%d %H:%M:%S", gmtime())
    mycode = '!./darknet detector train \"'+ dfg_file_path + "\" \"/content/gdrive/My Drive/darknet/yolov4_bees_new_training.cfg\" \"/content/gdrive/My Drive/darknet/bin4/yolov4.conv.137\" -dont_show -map"
    #print(mycode)
    exec(mycode)
    mts['train_end_time'] = strftime("%Y-%m-%d %H:%M:%S", gmtime())
```

## Train Model subsequently 


```
for mts in models_train_status:
  f = mts['cfg_file']
  dfg_file_path = ""+ f +""
  print("Train: ", dfg_file_path)
  print("Status: ", mts['train_status'])

  backup_folder_name = f[len(cfg_files_folder)+1:len(f)-5]
  backup_folder_path = '/content/gdrive/My Drive/yolo_models/backup_yolo4/on_qty_of_train_images/' + backup_folder_name + '/'
  backup_weights_file = backup_folder_path + 'yolov4_bees_new_training_last.weights'

  if mts['train_status'] == 'started':
    # from scratch 
    print("Start time: ", mts['train_start_time'])
    mts['train_status'] = 'started'
    mts['train_start_time'] = strftime("%Y-%m-%d %H:%M:%S", gmtime())
    mycode = '!./darknet detector train \"'+ dfg_file_path + "\" \"/content/gdrive/My Drive/darknet/yolov4_bees.cfg\" \"" + backup_weights_file + "\" -dont_show -map"
    #print(mycode)
    exec(mycode)
    mts['train_end_time'] = strftime("%Y-%m-%d %H:%M:%S", gmtime())
```
